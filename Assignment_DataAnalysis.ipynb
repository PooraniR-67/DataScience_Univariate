8)Test the similarity between the degree_t(Sci&Tech) and specialisation(
Mkt&HR) with respect to salary at significance level of 5%.(Make
decision using Hypothesis Testing)

```python
from scipy.stats import ttest_ind

group1 = dataset[dataset["degree_t"] == "Sci&Tech"]["salary"]
group2 = dataset[dataset["specialisation"] == "Mkt&HR"]["salary"]

t_stat, p_value = ttest_ind(group1, group2, equal_var=False)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
#In above case p = 0.01 < 0.05 (True) -> then we reject H0 null hypothesis and accept H1 alternate hypothesis
```

#10)What is the probability Density Function of the salary range from
700000 to 900000?

```python
from scipy.stats import norm

mean_salary = dataset["salary"].mean()
std_salary = dataset["salary"].std()

# Probability from 700000 to 900000
lower = norm.cdf(700000, mean_salary, std_salary)
upper = norm.cdf(900000, mean_salary, std_salary)

probability = upper - lower
print("Probability:", probability)
```

#11)Test the similarity between the degree_t(Sci&Tech)with respect to
etest_p and mba_p at significance level of 5%.(Make decision using Hypothesis
Testing)

```python
from scipy.stats import ttest_rel
sci = dataset[dataset["degree_t"] == "Sci&Tech"]
etest = sci["etest_p"]
mba = sci["mba_p"]

t_stat, p_value = ttest_rel(etest, mba)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
#In above case p > 0.05 , so accept Null hypotesting.
```
