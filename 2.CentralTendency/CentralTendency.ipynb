```python
from Univariate import Univariate
import pandas as pd
```

```python
dataset = pd.read_csv("Placement_Data_Full_Class.csv")
dataset.columns
```

```python
 def quanQual(dataset):
        qual = []
        quan = []
        for columnName in dataset.columns:
            if (dataset[columnName].dtype == 'O'):
                qual.append(columnName)
            else:
                quan.append(columnName)  
        return quan,qual
```

```python
quan,qual = quanQual(dataset)
print(quan,"\n",qual)
```

#Central Tendency



```python
#to find Mean, Median, Mode for a single column
mean = dataset["ssc_p"].mean()
median = dataset["ssc_p"].median()
mode = dataset["ssc_p"].mode()[0]
print(mean,"\n",median,"\n",mode)
```

```python
# For multiple column and to store the output values in a table

tbl_format = pd.DataFrame(index=["Mean", "Median", "Mode"], columns = quan)
tbl_format
```

```python
#To access particular value(cell) from the table
tbl_format["hsc_p"]["Mode"]
```

```python
#to assign the value for that cell
tbl_format["hsc_p"]["Mode"] = 90
tbl_format
```

```python
for columnName in quan:
    tbl_format[columnName]["Mean"] = dataset[columnName].mean()
    tbl_format[columnName]["Median"] = dataset[columnName].median()
    tbl_format[columnName]["Mode"] = dataset[columnName].mode()[0]

tbl_format
```
