```python
import pandas as pd
import numpy as np
dataset = pd.read_csv("sample_missing_values_dataset.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
from Univariate import Univariate
quan, qual = Univariate.quanQual(dataset)

from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imp.fit(dataset[quan])
processedtbl_quan = pd.DataFrame(imp.transform(dataset[quan]), columns = quan)
processedtbl_quan

imp = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
imp.fit(dataset[qual])
processedtbl_qual = pd.DataFrame(imp.transform(dataset[qual]), columns = qual)
processedtbl_qual

```

#Merge two tables in Pandas

```python
two_tables = [processedtbl_quan,processedtbl_qual]
preProcessed = pd.concat(two_tables, axis = 1)
preProcessed.to_csv("PreProcessedAssignmentData.csv")
```
